```python
# soal_02.py: Menampilkan grafik data kelulusan
def main():
    nilai = [float(input(f"Input nilai mahasiswa ke-{i+1}: ")) for i in range(10)]

    lulus = len([n for n in nilai if n >= 60])
    gagal = 10 - lulus

    print(f"\nJumlah Lulus: {lulus}")
    print(f"Jumlah Gagal: {gagal}")

    print("\nGrafik Kelulusan:")
    print(f"Lulus ({lulus}) : {'█' * lulus}")
    print(f"Gagal ({gagal}) : {'█' * gagal}")

if __name__ == "__main__":
    main()
```
