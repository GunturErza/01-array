```python
# Soal_01.py: Menampilkan grafik nilai tertinggi dan terendah
def main():
    nilai = [float(input(f"Input nilai mahasiswa ke-{i+1}: ")) for i in range(10)]

    tertinggi = max(nilai)
    terendah = min(nilai)

    print(f"\nNilai Tertinggi: {tertinggi}")
    print(f"Nilai Terendah : {terendah}")

    # Visualisasi sederhana menggunakan karakter
    print("\nGrafik Nilai Tertinggi & Terendah:")
    print(f"Tertinggi ({tertinggi}): {'█' * int(tertinggi // 5)}")
    print(f"Terendah  ({terendah}) : {'█' * int(terendah // 5)}")

if __name__ == "__main__":
    main()
```
